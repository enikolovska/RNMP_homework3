```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os
```

```python
data = './data/diabetes_binary_health_indicators_BRFSS2015.csv'
```

```python
df = pd.read_csv(data)
```

```python
print(f"Number of rows: {df.shape[0]:,}")
print(f"Number of columns: {df.shape[1]}")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
print(df['Diabetes_binary'].value_counts())
print('\nPercentage of Diabetes Binary', df['Diabetes_binary'].value_counts() / df.shape[0] * 100)
```

```python
offline_data, online_data = train_test_split(df,test_size=0.2, random_state=43, stratify=df['Diabetes_binary'])
```

```python
print('Ofline data shape:', offline_data.shape)
print('Online data shape:', online_data.shape)
```

```python
print('Offline class counts \n',offline_data['Diabetes_binary'].value_counts())
print('\nOnline class counts \n', online_data['Diabetes_binary'].value_counts())
```

```python
os.makedirs('./data', exist_ok=True)

offline_path = './data/offline.csv'
online_path = './data/online.csv'

offline_data.to_csv(offline_path, index=False)
online_data.to_csv(online_path, index=False)
```
